```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clv_data=pd.read_csv(r'E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_dataset.csv')
```

```python
clv_data.head()
```

```python
clv_data.shape
```

```python
X=clv_data.drop(columns=['CustomerID','FutureRevenue','FutureRevenue_log'])
Y=clv_data['FutureRevenue_log']
```

```python
X,Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
linear_reg=LinearRegression()
linear_reg.fit(x_train,y_train)
```

```python
linear_reg_pred=linear_reg.predict(x_test)
```

```python
linear_reg_pred
```

```python
mae_lr=mean_absolute_error(y_test,linear_reg_pred)
rmse_lr=np.sqrt(mean_squared_error(y_test,linear_reg_pred))
r2_lr=r2_score(y_test,linear_reg_pred)
```

```python
print("Mean Absolute Error:", mae_lr)
print("Root Mean Squared Error:", rmse_lr)
print("R2 Score:", r2_lr)
```

```python
output_path = r"E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\lr_model_results.txt"

with open(output_path, "w") as f:
    f.write("Linear Regression Results\n")
    f.write(f"Mean Absolute Error: {mae_lr}\n")
    f.write(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test, linear_reg_pred))}\n")
    f.write(f"R2 Score: {r2_lr}\n")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(y_test,linear_reg_pred, alpha=0.5)
plt.xlabel("Actual CLV (log)")
plt.ylabel("Predicted CLV (log)")
plt.title("Linear Regression: Actual vs Predicted")

plt.savefig(r'E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\lr_actual_vs_pred.png')
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_for=RandomForestRegressor(n_estimators=200,random_state=42)
random_for.fit(x_train,y_train)
```

```python
random_for_pred=random_for.predict(x_test)
```

```python
mae_rf=mean_absolute_error(y_test,random_for_pred)
rmse_rf=np.sqrt(mean_squared_error(y_test,random_for_pred))
r2_rf=r2_score(y_test,random_for_pred)
```

```python
print("Mean Absolute Error:", mae_rf)
print("Root Mean Squared Error:", rmse_rf)
print("R2 Score:", r2_rf)
```

```python
output_path = r"E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\rf_model_results.txt"

with open(output_path, "w") as f:
    f.write("Random Forest Results\n")
    f.write(f"Mean Absolute Error: {mae_rf}\n")
    f.write(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test, random_for_pred))}\n")
    f.write(f"R2 Score: {r2_rf}\n")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(y_test,random_for_pred, alpha=0.5)
plt.xlabel("Actual CLV (log)")
plt.ylabel("Predicted CLV (log)")
plt.title("Random Forest: Actual vs Predicted")

plt.savefig(r'E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\rf_actual_vs_pred.png')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
grad_boost=GradientBoostingRegressor(random_state=42)
grad_boost.fit(x_train,y_train)
```

```python
grad_boost_pred=grad_boost.predict(x_test)
```

```python
mae_gb=mean_absolute_error(y_test,grad_boost_pred)
rmse_gb=np.sqrt(mean_squared_error(y_test,grad_boost_pred))
r2_gb=r2_score(y_test,grad_boost_pred)
```

```python
print("Mean Absolute Error:", mae_gb)
print("Root Mean Squared Error:", rmse_gb)
print("R2 Score:", r2_gb)
```

```python
output_path = r"E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\xg_model_results.txt"

with open(output_path, "w") as f:
    f.write("Gradient Boosting Results\n")
    f.write(f"Mean Absolute Error: {mae_gb}\n")
    f.write(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test, grad_boost_pred))}\n")
    f.write(f"R2 Score: {r2_gb}\n")
```

```python
plt.figure(figsize=(6,4))
plt.scatter(y_test,grad_boost_pred, alpha=0.5)
plt.xlabel("Actual CLV (log)")
plt.ylabel("Predicted CLV (log)")
plt.title("Gradient Boosting: Actual vs Predicted")

plt.savefig(r'E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\xg_actual_vs_pred.png')
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest", "Gradient Boosting"],
    "MAE": [mae_lr, mae_rf, mae_gb],
    "RMSE": [rmse_lr, rmse_rf, rmse_gb],
    "R2": [r2_lr, r2_rf, r2_gb]
})

results.sort_values("R2", ascending=False)
```

```python
results.to_csv(r'E:\SEM 8\Sales_Intelligence_System\Code_Implementation\Online_retail\clv_modeling\clv_model_comparison.csv',index=False)
```
